```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pokemon = pd.read_csv('pokemon_data.csv')
pokemon.head()
```

```python
pokemon.describe()
```

```python
pokemon['Type 1'].value_counts().plot(kind='pie', autopct='%1.1f%%', cmap='tab20c', figsize=(10, 8))
```

```python
pokemon['Total'] = pokemon[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].sum(axis=1)
pokemon.head()
```

```python
list(pokemon.columns)
```

```python
pokemon = pokemon[['#', 'Name', 'Type 1', 'Type 2', 'Total', 'HP', 'Attack', 'Defense', 'Sp. Atk',
       'Sp. Def', 'Speed', 'Generation', 'Legendary']]
pokemon.head()
```

```python
pokemon['Total'].plot(kind='hist', figsize=(10, 8))
```

```python
pokemon['Total'].plot(kind='box', vert=False, figsize=(10, 5))
```

```python
pokemon['Legendary'].value_counts().plot(kind='pie', autopct='%1.1f%%', cmap='Set3', figsize=(10, 8))
```

```python
sns.boxplot(data=pokemon, x='Attack')
```

```python
pokemon.loc[pokemon['Attack'] > 165].sort_values('Attack', ascending=False)
```

```python
sns.boxplot(data=pokemon, x='Speed')
```

```python
ice_pokemon_max_defense = pokemon.loc[pokemon['Type 1'] == 'Ice', 'Defense'].max()
pokemon.loc[(pokemon['Type 1'] == 'Ice') & (pokemon['Defense'] == ice_pokemon_max_defense)]
```

```python
pokemon.loc[pokemon['Legendary'], 'Type 1'].value_counts()
```

```python
pokemon.loc[pokemon['Legendary'], 'Type 1'].value_counts().plot(kind='bar')
```

```python
pokemon['Generation'].value_counts(sort=False)
```

```python
pokemon.loc[(pokemon['Type 1'] == 'Water') & (pokemon['Generation'].isin({1, 2, 3}))].sort_values(['Total'], ascending=False).head()
```

```python
pokemon.loc[
    ((pokemon['Type 1'] == 'Dragon') | (pokemon['Type 2'] == 'Dragon')) 
    & (pokemon['Generation'].isin({5, 6}))
].sort_values(['Total'], ascending=False).head()
```

```python
pokemon.loc[
    (pokemon['Type 1'] == 'Fire')
    & (pokemon['Legendary']),
    ['Name', 'Attack', 'Generation']
]
```

```python
speed = pokemon['Speed'].plot(kind='hist', figsize=(10, 5), bins=100)
speed.axvline(pokemon['Speed'].quantile(.05), color='red')
speed.axvline(pokemon['Speed'].quantile(.95), color='red')
```

```python
bottom_5 = pokemon['Speed'].quantile(.05)
top_5 = pokemon['Speed'].quantile(.95)
(bottom_5, top_5)
```

```python
pokemon.query('Speed < @bottom_5 or Speed > @top_5').sample(5)
```

```python
fig, ax = plt.subplots(figsize=(14, 7))
sns.scatterplot(data=pokemon, x="Defense", y="Attack", hue='Legendary', ax=ax)
ax.annotate(
    "Who's this pokemon?", xy=(160, 180), xytext=(160, 150), color='red',
    arrowprops=dict(arrowstyle="->", color='red')
)
```
